```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
def threebody_simprocess_new(filename,m1,m2,m3):
    dat = np.loadtxt(filename)
    G = 6.6743*1.98892*3.1536*3.1536/(1.496*1.496*1.496)
    t = dat[:,0]
    x1 = dat[:,1]
    y1 = dat[:,2]
    x2 = dat[:,3]
    y2 = dat[:,4]
    x3 = dat[:,5]
    y3 = dat[:,6]
    v_x1 = dat[:,7]
    v_y1 = dat[:,8]
    v_x2 = dat[:,9]
    v_y2 = dat[:,10]
    v_x3 = dat[:,11]
    v_y3 = dat[:,12]
    r12 = np.sqrt((x1-x2)**2 + (y1-y2)**2)
    r13 = np.sqrt((x1-x3)**2 + (y1-y3)**2)
    r23 = np.sqrt((x2-x3)**2 + (y2-y3)**2)
    E_kin_total = m1*(v_x1**2 + v_y1**2)/2 + m2*(v_x2**2 + v_y2**2)/2 + m3*(v_x3**2 + v_y3**2)/2 
    E_pot_total = - G*m1*m2/r12 - G*m1*m3/r13 - G*m2*m3/r23

    #plots:
    fs = 15
    plt.figure(figsize = (12,7))
    plt.subplot(2,2,1)
    plt.plot(x1,y1,label ="body 1")
    plt.plot(x2,y2,label ="body 2")
    plt.plot(x3,y3,label ="body 3")
    plt.xlabel("x [AU]",fontsize = fs)
    plt.ylabel("y [AU]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)

    plt.subplot(2,2,2)
    plt.plot(v_x1,v_y1,label ="body 1")
    plt.plot(v_x2,v_y2,label ="body 2")
    plt.plot(v_x3,v_y3,label ="body 3")
    plt.xlabel("$v_x$ [AU/yr]",fontsize = fs)
    plt.ylabel("$v_y$ [AU/yr]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)

    plt.subplot(2,2,3)
    plt.plot(t,r12,"b",label ="body 1&2")
    plt.plot(t,r13,"r",label ="body 1&3")
    plt.plot(t,r23,"y",label ="body 2&3")
    plt.xlabel("t [yr]",fontsize = fs)
    plt.ylabel("|$r_i - r_j$| [AU]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)
    print(min(r23))
    plt.subplot(2,2,4)
    plt.plot(t,E_pot_total+E_kin_total)
    plt.xlabel("t [yr]",fontsize = fs)
    plt.ylabel("$E_{total}$ [AU]",fontsize = fs)
    plt.grid()

    plt.tight_layout()
    print(dat[dat.shape[1]-1,1:])
    print(dat[0,1:])
```

```python
def threeplus1body_simprocess_new_balls(filename,m1,m2,m3):
    dat = np.loadtxt(filename)
    G = 6.6743*1.98892*3.1536*3.1536/(1.496*1.496*1.496)
    t = dat[:,0]
    x1 = dat[:,1]
    y1 = dat[:,2]
    x2 = dat[:,3]
    y2 = dat[:,4]
    x3 = dat[:,5]
    y3 = dat[:,6]
    v_x1 = dat[:,7]
    v_y1 = dat[:,8]
    v_x2 = dat[:,9]
    v_y2 = dat[:,10]
    v_x3 = dat[:,11]
    v_y3 = dat[:,12]
    x4 = dat[:,13]
    y4 = dat[:,14]
    v_x4 = dat[:,15]
    v_y4 = dat[:,16]
    r12 = np.sqrt((x1-x2)**2 + (y1-y2)**2)
    r13 = np.sqrt((x1-x3)**2 + (y1-y3)**2)
    r23 = np.sqrt((x2-x3)**2 + (y2-y3)**2)
    
    r14 = np.sqrt((x1-x4)**2 + (y1-y4)**2)
    r24 = np.sqrt((x2-x4)**2 + (y2-y4)**2)
    r34 = np.sqrt((x3-x4)**2 + (y3-y4)**2)
    
    E_kin_total = m1*(v_x1**2 + v_y1**2)/2 + m2*(v_x2**2 + v_y2**2)/2 + m3*(v_x3**2 + v_y3**2)/2
    E_pot_total = - G*m1*m2/r12 - G*m1*m3/r13 - G*m2*m3/r23

    #plots:
    fs = 15
    plt.figure(figsize = (12,11))
    plt.subplot(3,2,1)
    plt.plot(x1,y1,"b",label ="body 1")
    plt.plot(x2,y2,color="orange",label ="body 2")
    plt.plot(x3,y3,"g",label ="body 3")
    plt.plot(x4,y4,color = "black",label ="body 4")
    plt.scatter([x1[-1]], [y1[-1]], color="b", s=50)
    plt.scatter([x2[-1]], [y2[-1]], color="orange", s=50)
    plt.scatter([x3[-1]], [y3[-1]], color="g", s=50)
    plt.scatter([x4[-1]], [y4[-1]], color="black", s=20)
    plt.xlabel("x [AU]",fontsize = fs)
    plt.ylabel("y [AU]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)

    plt.subplot(3,2,2)
    plt.plot(v_x1,v_y1,label ="body 1")
    plt.plot(v_x2,v_y2,label ="body 2")
    plt.plot(v_x3,v_y3,label ="body 3")
    plt.plot(v_x4,v_y4,label ="body 4")
    plt.xlabel("$v_x$ [AU/yr]",fontsize = fs)
    plt.ylabel("$v_y$ [AU/yr]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)

    plt.subplot(3,2,3)
    plt.plot(t,r12,"b",label ="body 1&2")
    plt.plot(t,r13,"r",label ="body 1&3")
    plt.plot(t,r23,"y",label ="body 2&3")
    plt.xlabel("t [yr]",fontsize = fs)
    plt.ylabel("|$r_i - r_j$| [AU]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)
    print(min(r23))
    plt.subplot(3,2,4)
    plt.plot(t,E_pot_total+E_kin_total)
    plt.xlabel("t [yr]",fontsize = fs)
    plt.ylabel("$E_{total}$ [a.u.]",fontsize = fs)
    plt.grid()
    
    plt.subplot(3,2,5)
    plt.plot(t,r14,"b",label ="body 1&4")
    #plt.plot(t,r24,"r",label ="body 2&4")
    #plt.plot(t,r34,"y",label ="body 3&4")
    plt.xlabel("t [yr]",fontsize = fs)
    plt.ylabel("|$r_i - r_j$| [AU]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)
    
    plt.subplot(3,2,6)
    plt.plot(t,r14,"b",label ="body 1&4")
    plt.plot(t,r24,"r",label ="body 2&4")
    plt.plot(t,r34,"y",label ="body 3&4")
    plt.xlabel("t [yr]",fontsize = fs)
    plt.ylabel("|$r_i - r_j$| [AU]",fontsize = fs)
    plt.grid()
    plt.legend(loc = "best",fontsize = fs-2)

    plt.tight_layout()
    print(dat[dat.shape[1]-1,1:])
    print(dat[0,1:])
```

```python
def downsample(data,M):
    res = np.zeros((int(data.shape[0]/M)))
    for i in range(res.shape[0]):
        res[i] = data[i*M]
    return res

def CM(m1,m2,m3,x1,x2,x3,y1,y2,y3):
    return (m1*x1+m2*x2+m3*x3)/(m1+m2+m3), (m1*y1+m2*y2+m3*y3)/(m1+m2+m3)
```

```python
filename="build/Debug/3body-VV2_myparams_videotest.dat"
dat = np.loadtxt(filename)
G = 6.6743*1.98892*3.1536*3.1536/(1.496*1.496*1.496)
M = 40
t = downsample(dat[:,0],M)
x1 = downsample(dat[:,1],M)
y1 = downsample(dat[:,2],M)
x2 = downsample(dat[:,3],M)
y2 = downsample(dat[:,4],M)
x3 = downsample(dat[:,5],M)
y3 = downsample(dat[:,6],M)

cm_x,cm_y = CM(m1,m2,m3,x1,x2,x3,y1,y2,y3)

x1 += - cm_x
x2 += - cm_x
x3 += - cm_x
y1 += - cm_y
y2 += - cm_y
y3 += - cm_y

min_x = np.array([min(x1),min(x2),min(x3)])
min_y = np.array([min(y1),min(y2),min(y3)])

max_x = np.array([max(x1),max(x2),max(x3)])
max_y = np.array([max(y1),max(y2),max(y3)])

minxy=min(np.array([min(np.array([min(x1),min(x2),min(x3)])),min(np.array([min(y1),min(y2),min(y3)]))]))
maxxy=max(np.array([max(np.array([max(x1),max(x2),max(x3)])),max(np.array([max(y1),max(y2),max(y3)]))]))
```

```python
fs = 15
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, autoscale_on=False, xlim=(minxy, maxxy), ylim=(minxy, maxxy))
ax.set_xlabel("X [AU]", fontsize = fs+2)
ax.set_ylabel("Y [AU]", fontsize = fs+2)
ax.set_title("2D 3 body - Velocity-Verlet, $m_1 = 3, m_2 = 2, m_3 = 5$, h = 0.001", fontsize = fs-3)

R = 0.5

patch1 = plt.Circle((5, -5), R, fc='blue')
patch2 = plt.Circle((5, -5), R, fc='orange')
patch3 = plt.Circle((5, -5), R, fc='green')
   
# data which the line will  
# contain (x, y) 
def init():  
    ax.add_patch(patch1)
    ax.add_patch(patch2)
    ax.add_patch(patch3)
    return patch1,patch2,patch3
   
def animate(i): 
    patch1.center = (x1[i], y1[i])
    patch2.center = (x2[i], y2[i])
    patch3.center = (x3[i], y3[i])
    return patch1,patch2,patch3
   
anim = FuncAnimation(fig, animate, init_func = init, 
                     frames = 3750, interval = 20, blit = True) 
  
   
anim.save('threebody2DVV_myparams.mp4',  
          writer = 'ffmpeg', fps = 50)
```
