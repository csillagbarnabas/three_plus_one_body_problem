```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat = np.loadtxt("build/Debug/2body-test.dat")
```

```python
G = 6.6743*1.98892*3.1536*3.1536/(1.496*1.496*1.496)
m1 = 3
m2 = 2 #metadata
```

```python
dat.shape
```

```python
t = dat[:,0]
x1 = dat[:,1]
y1 = dat[:,2]
v_x1 = dat[:,3]
v_y1 = dat[:,4]
x2 = dat[:,5]
y2 = dat[:,6]
v_x2 = dat[:,7]
v_y2 = dat[:,8]
E_kin_total = m1*(v_x1**2 + v_y1**2)/2 + m2*(v_x2**2 + v_y2**2)/2
E_pot_total = - G*m1*m2/np.sqrt((x1-x2)**2 + (y1-y2)**2)
```

```python
plt.plot(x1,y1)
```

```python
plt.plot(x2,y2)
```

```python
plt.plot(v_x1,v_y1)
```

```python
plt.plot(v_x2,v_y2)
```

```python
plt.plot(t,E_kin_total)
```

```python
plt.plot(t,E_pot_total)
```

```python
plt.plot(t,np.sqrt((x1-x2)**2 + (y1-y2)**2))
```

```python
plt.plot(t,E_pot_total+E_kin_total)
```
