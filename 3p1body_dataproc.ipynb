```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat = np.loadtxt("build/Debug/2body-test.dat")

G = 6.6743*1.98892*3.1536*3.1536/(1.496*1.496*1.496)
m1 = 3
m2 = 2 #metadata

t = dat[:,0]
x1 = dat[:,1]
y1 = dat[:,2]
v_x1 = dat[:,3]
v_y1 = dat[:,4]
x2 = dat[:,5]
y2 = dat[:,6]
v_x2 = dat[:,7]
v_y2 = dat[:,8]

r12 = np.sqrt((x1-x2)**2 + (y1-y2)**2)
E_kin_total = m1*(v_x1**2 + v_y1**2)/2 + m2*(v_x2**2 + v_y2**2)/2
E_pot_total = - G*m1*m2/r12
```

```python
dat = np.loadtxt("build/Debug/3body-test.dat")

G = 6.6743*1.98892*3.1536*3.1536/(1.496*1.496*1.496)
m1 = 3
m2 = 2 
m3 = 5 #metadata


t = dat[:,0]
x1 = dat[:,1]
y1 = dat[:,2]
v_x1 = dat[:,3]
v_y1 = dat[:,4]

x2 = dat[:,5]
y2 = dat[:,6]
v_x2 = dat[:,7]
v_y2 = dat[:,8]

x3 = dat[:,9]
y3 = dat[:,10]
v_x3 = dat[:,11]
v_y3 = dat[:,12]

r12 = np.sqrt((x1-x2)**2 + (y1-y2)**2)
r13 = np.sqrt((x1-x3)**2 + (y1-y3)**2)
r23 = np.sqrt((x2-x3)**2 + (y2-y3)**2)
E_kin_total = m1*(v_x1**2 + v_y1**2)/2 + m2*(v_x2**2 + v_y2**2)/2 + m3*(v_x3**2 + v_y3**2)/2 
E_pot_total = - G*m1*m2/r12 - G*m1*m3/r13 - G*m2*m3/r23
```

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
```

```python
plt.plot(v_x1,v_y1)
plt.plot(v_x2,v_y2)
plt.plot(v_x3,v_y3)
```

```python
plt.plot(t,E_kin_total)
```

```python
plt.plot(t,E_pot_total)
```

```python
plt.plot(t,r12)
plt.plot(t,r13)
plt.plot(t,r23)
```

```python
plt.plot(t,E_pot_total+E_kin_total)
```
